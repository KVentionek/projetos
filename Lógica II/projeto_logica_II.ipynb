```python
def convert(chave, indice):
  for i, x in enumerate(ler):
    alunos[i][chave]=ler[i][indice]

import csv

with open('alunos_matric.csv', mode='r') as arquivo:
    ler = list(csv.reader(arquivo))
    chaves = ler[0]

alunos=[{'Nome': ler[i][0]} for i, x in enumerate(ler)]

convert('CPF', 1)
convert('Matrícula', 2)
convert('Notas (Prova 1)', 3)
convert('Notas (Prova 2)', 4)
convert('Notas (Prova 3)', 5)
convert('Notas (Prova 4)', 6)
convert('Notas (Prova 5)', 7)
convert('Rematrícula', 8)
convert('Faltas', 9)
convert('Telefone', 10)
convert('Mensalidades em atraso', 11)
convert('Curso', 12)

alunos.pop(0) #aqui estou removendo a linha que ficou apenas o cabeçalho como chave e item

def exibir_dados():
  for aluno in alunos:
      print(f"\nDados do aluno:\nNome: {aluno['Nome']}\nNúmero de Matrícula: {aluno['Matrícula']}\nTelefone: {aluno['Telefone']}")


def pesquisar_aluno(mat_procurado):
  procurar_aluno = {d['Matrícula']: d for d in alunos}

  try:
    aluno_procurado = procurar_aluno[mat_procurado]
    print(f"\nDados do aluno:\nNome: {aluno_procurado['Nome']}\nCPF: {aluno_procurado['CPF']}\nNúmero de Matrícula: {aluno_procurado['Matrícula']}\nTelefone: {aluno_procurado['Telefone']}\nRematrícula: {aluno_procurado['Rematrícula']}\nFaltas: {aluno_procurado['Faltas']}\nMensalidades em atraso: {aluno_procurado['Mensalidades em atraso']}")
    print(f"\nAvaliações: \nNotas (Prova 1): {aluno_procurado['Notas (Prova 1)']}\nNotas (Prova 2): {aluno_procurado['Notas (Prova 2)']}\nNotas (Prova 3): {aluno_procurado['Notas (Prova 3)']}\nNotas (Prova 4): {aluno_procurado['Notas (Prova 4)']}\nNotas (Prova 5): {aluno_procurado['Notas (Prova 5)']}")

  except KeyError:
      print("Aluno não encontrado.")
  except Exception as e:
      print("Ocorreu um erro:", str(e))


# Função para calcular a média das notas de um aluno
def calcular_media(aluno):
    notas = [float(aluno[f'Notas (Prova {i})']) for i in range(1, 6)]
    return sum(notas) / len(notas)


def exibir_todas_notas():
  notas_alunos = []
  for i, x in enumerate(alunos):

    nota_aluno =[]

    nota_aluno.append(alunos[i]['Notas (Prova 1)'] )
    nota_aluno.append(alunos[i]['Notas (Prova 2)'] )
    nota_aluno.append(alunos[i]['Notas (Prova 3)'] )
    nota_aluno.append(alunos[i]['Notas (Prova 4)'] )
    nota_aluno.append(alunos[i]['Notas (Prova 5)'] )
    aluno_nota = {alunos[i]['Nome']: nota_aluno}
    notas_alunos.append(aluno_nota)

  for aluno in notas_alunos:
      nome = next(iter(aluno))
      notas = aluno[nome]
      print(f"\nNotas do aluno: {nome}")
      for i, nota in enumerate(notas, start=1):
          print(f"Nota Prova {i}: {nota}")

def encontrar_nota_aluno(opcao):
  try:
      encontrado = False
      for aluno in alunos:
          if opcao in aluno['Nome'] or opcao == aluno['Matrícula']:
              encontrado = True
              print(f"\nNotas do aluno: {aluno['Nome']}")
              print(f"Nota Prova 1: {aluno['Notas (Prova 1)']}")
              print(f"Nota Prova 2: {aluno['Notas (Prova 2)']}")
              print(f"Nota Prova 3: {aluno['Notas (Prova 3)']}\nNota Prova 4: {aluno['Notas (Prova 4)']}\nNota Prova 5: {aluno['Notas (Prova 5)']}\n")
              break

      if not encontrado:
          print("Aluno não encontrado.")
  except KeyError:
      print("Chave inválida no dicionário de alunos. Verifique a estrutura dos dados.")
      opcao = input('Informe o nome ou número de matrícula do aluno que deseja encontrar: ').capitalize()


# Função para encontrar os alunos com a maior e menor nota em uma prova específica
def encontrar_maior_e_menor_nota_por_prova(prova):
    notas_prova = [float(aluno[f'Notas (Prova {prova})']) for aluno in alunos]
    maior_nota = max(notas_prova)
    menor_nota = min(notas_prova)
    alunos_maior_nota = [aluno['Nome'] for aluno in alunos if float(aluno[f'Notas (Prova {prova})']) == maior_nota]
    alunos_menor_nota = [aluno['Nome'] for aluno in alunos if float(aluno[f'Notas (Prova {prova})']) == menor_nota]
    return maior_nota, menor_nota, alunos_maior_nota, alunos_menor_nota

# Encontrar os alunos com a maior e menor nota em cada prova
def alunos_ranking_provas():
  resultados_por_prova = {}

  for prova in range(1, 6):
      maior_nota_prova, menor_nota_prova, alunos_maior_nota_prova, alunos_menor_nota_prova = encontrar_maior_e_menor_nota_por_prova(prova)
      resultados_por_prova[prova] = {'Maior Nota': maior_nota_prova, 'Menor Nota': menor_nota_prova, 'Alunos com Maior Nota': alunos_maior_nota_prova, 'Alunos com Menor Nota': alunos_menor_nota_prova}

  # Encontrar o aluno com a maior e menor média nas provas
  maior_media = None
  menor_media = None
  aluno_maior_media = None
  aluno_menor_media = None

  for aluno in alunos:
      media = calcular_media(aluno)
      if maior_media is None or media > maior_media:
          maior_media = media
          aluno_maior_media = aluno['Nome']
      if menor_media is None or media < menor_media:
          menor_media = media
          aluno_menor_media = aluno['Nome']

  # Dicionário para armazenar os resultados por curso
  resultados_por_curso = {}

  # Calcular as maiores e menores notas e médias por curso
  for aluno in alunos:
      curso = aluno['Curso']
      media = calcular_media(aluno)

      if curso not in resultados_por_curso:
          resultados_por_curso[curso] = {
              'Maior Nota': None,
              'Menor Nota': None,
              'Alunos com Maior Nota': [],
              'Alunos com Menor Nota': [],
              'Maior Média': None,
              'Menor Média': None,
              'Alunos com Maior Média': [],
              'Alunos com Menor Média': []
          }

      # Verificar se é a maior ou menor nota
      for prova in range(1, 6):
          nota = float(aluno[f'Notas (Prova {prova})'])
          if resultados_por_curso[curso]['Maior Nota'] is None or nota > resultados_por_curso[curso]['Maior Nota']:
              resultados_por_curso[curso]['Maior Nota'] = nota
              resultados_por_curso[curso]['Alunos com Maior Nota'] = [aluno['Nome']]
          elif nota == resultados_por_curso[curso]['Maior Nota']:
              resultados_por_curso[curso]['Alunos com Maior Nota'].append(aluno['Nome'])

          if resultados_por_curso[curso]['Menor Nota'] is None or nota < resultados_por_curso[curso]['Menor Nota']:
              resultados_por_curso[curso]['Menor Nota'] = nota
              resultados_por_curso[curso]['Alunos com Menor Nota'] = [aluno['Nome']]
          elif nota == resultados_por_curso[curso]['Menor Nota']:
              resultados_por_curso[curso]['Alunos com Menor Nota'].append(aluno['Nome'])

      # Verificar se é a maior ou menor média
      if resultados_por_curso[curso]['Maior Média'] is None or media > resultados_por_curso[curso]['Maior Média']:
          resultados_por_curso[curso]['Maior Média'] = media
          resultados_por_curso[curso]['Alunos com Maior Média'] = [aluno['Nome']]
      elif media == resultados_por_curso[curso]['Maior Média']:
          resultados_por_curso[curso]['Alunos com Maior Média'].append(aluno['Nome'])

      if resultados_por_curso[curso]['Menor Média'] is None or media < resultados_por_curso[curso]['Menor Média']:
          resultados_por_curso[curso]['Menor Média'] = media
          resultados_por_curso[curso]['Alunos com Menor Média'] = [aluno['Nome']]
      elif media == resultados_por_curso[curso]['Menor Média']:
          resultados_por_curso[curso]['Alunos com Menor Média'].append(aluno['Nome'])

  # Imprimir resultados por prova
  print("Resultados por prova:")
  for prova, resultados in resultados_por_prova.items():
      print(f"Prova {prova}")
      print(f"Maior Nota: {resultados['Maior Nota']}")
      print(f"Alunos com Maior Nota: {resultados['Alunos com Maior Nota']}")
      print(f"Menor Nota: {resultados['Menor Nota']}")
      print(f"Alunos com Menor Nota: {resultados['Alunos com Menor Nota']}")
      print()

  # Imprimir resultados de maior e menor média
  print(f"Resultado de maior e menor média:\nMaior Média: {aluno_maior_media}\nMenor Média: {aluno_menor_media}\n")

  # Imprimir resultados por curso
  print("Resultados por curso:")
  for curso, resultados in resultados_por_curso.items():
      print(f"Curso: {curso}")
      print(f"Maior Nota: {resultados['Maior Nota']}")
      print(f"Alunos com Maior Nota: {resultados['Alunos com Maior Nota']}")
      print(f"Menor Nota: {resultados['Menor Nota']}")
      print(f"Alunos com Menor Nota: {resultados['Alunos com Menor Nota']}")
      print(f"Maior Média: {resultados['Maior Média']}")
      print(f"Alunos com Maior Média: {resultados['Alunos com Maior Média']}")
      print(f"Menor Média: {resultados['Menor Média']}")
      print(f"Alunos com Menor Média: {resultados['Alunos com Menor Média']}")
      print()


#essa função exibe todas as médias
def exibir_todas_medias ():
  for aluno in alunos:
    notas = []
    for chave, valor in aluno.items():
      if chave.startswith('Notas (Prova'):
         notas.append(float(valor))
    media = sum(notas) / len(notas)
    aluno['Média das Notas'] = media

  for aluno in alunos:
    print(f"Aluno: {aluno['Nome']} \nMédia das Notas: {aluno['Média das Notas']:.1f}\n")

# Função para encontrar o aluno com a maior e menor nota em um determinado curso
def encontrar_maior_e_menor_nota_por_curso(curso):
    alunos_curso = [aluno for aluno in alunos if aluno['Curso'] == curso]
    maior_nota = None
    menor_nota = None
    aluno_maior_nota = None
    aluno_menor_nota = None

    for aluno in alunos_curso:
        media = calcular_media(aluno)
        if maior_nota is None or media > maior_nota:
            maior_nota = media
            aluno_maior_nota = aluno['Nome']
        if menor_nota is None or media < menor_nota:
            menor_nota = media
            aluno_menor_nota = aluno['Nome']

    return aluno_maior_nota, aluno_menor_nota

def encontrar_medias():
  # Encontrar o aluno com a maior e menor nota no geral
  maior_nota_geral = None
  menor_nota_geral = None
  aluno_maior_nota_geral = None
  aluno_menor_nota_geral = None

  for aluno in alunos:
      media = calcular_media(aluno)
      if maior_nota_geral is None or media > maior_nota_geral:
          maior_nota_geral = media
          aluno_maior_nota_geral = aluno['Nome']
      if menor_nota_geral is None or media < menor_nota_geral:
          menor_nota_geral = media
          aluno_menor_nota_geral = aluno['Nome']

  # Encontrar o aluno com a maior e menor nota em cada curso
  cursos = set(aluno['Curso'] for aluno in alunos)
  resultados_por_curso = {}

  for curso in cursos:
      aluno_maior_nota_curso, aluno_menor_nota_curso = encontrar_maior_e_menor_nota_por_curso(curso)
      resultados_por_curso[curso] = {'Maior Nota': aluno_maior_nota_curso, 'Menor Nota': aluno_menor_nota_curso}

  # Resultados gerais
  resultado_geral = {'Maior Nota': aluno_maior_nota_geral, 'Menor Nota': aluno_menor_nota_geral}

  # Imprimir resultados
  print("Resultados por curso:")
  for curso, resultados in resultados_por_curso.items():
      print(f"Curso: {curso}\nMaior Nota: {resultados['Maior Nota']}\nMenor Nota: {resultados['Menor Nota']}")

  print(f"Resultado Geral:\nMaior Nota: {resultado_geral['Maior Nota']}\nMenor Nota: {resultado_geral['Menor Nota']}")

def medias_iguais():
  soma_notas = 0
  num_notas = 0

  for aluno in alunos:
      for chave, valor in aluno.items():
          if chave.startswith('Notas (Prova'):
              soma_notas += float(valor)
              num_notas += 1

  media_geral = soma_notas / num_notas

  print(f'Média Geral das Notas: {media_geral:.1f}')
  print()

  # Verificar se houve médias repetidas entre os alunos
  medias = [aluno['Média das Notas'] for aluno in alunos]
  medias_repetidas = []

  for media in medias:
      if medias.count(media) > 1 and media not in medias_repetidas:
          medias_repetidas.append(media)

  if len(medias_repetidas) > 0:
      print("Médias Repetidas:")
      for media in medias_repetidas:
          print(f'Média: {media:.1f}')
          alunos_com_media_repetida = [aluno['Nome'] for aluno in alunos if aluno['Média das Notas'] == media]
          print(f'Alunos: {alunos_com_media_repetida}\n')
  else:
      print("Não há Médias Repetidas entre os Alunos")

# Função para calcular o valor em aberto de um aluno
def calcular_valor(aluno):
    mensalidades_em_atraso = int(aluno['Mensalidades em atraso'])
    return mensalidades_em_atraso * 100

def arquivo_financeiro():
  header = ['Nome do Aluno', 'Mensalidades em Atraso', 'Valor em Aberto']

  # Lista para armazenar os dados de cada aluno
  dados_alunos = []

  # Calcular os dados de cada aluno
  for aluno in alunos:
      nome_aluno = aluno['Nome']
      mensalidades_atraso = int(aluno['Mensalidades em atraso'])
      valor_em_aberto = calcular_valor(aluno)
      dados_alunos.append([nome_aluno, mensalidades_atraso, valor_em_aberto])

  # Calcular o total de mensalidades em atraso e o total do valor em aberto
  total_mensalidades_atraso = sum(int(aluno['Mensalidades em atraso']) for aluno in alunos)
  total_valor_em_aberto = sum(calcular_valor(aluno) for aluno in alunos)

  # Adicionar linha com os totais no final do arquivo
  dados_alunos.append(['Total', total_mensalidades_atraso, total_valor_em_aberto])

  # Escrever os dados no arquivo CSV
  with open('financeiro_alunos.csv', 'w', newline='') as arquivo_csv:
      writer = csv.writer(arquivo_csv)
      writer.writerow(header)
      writer.writerows(dados_alunos)

  print("Arquivo CSV gerado com sucesso!")


# Função para calcular a porcentagem de faltas
def calcular_porcentagem_presenca(aluno):
    faltas = int(aluno['Faltas'])
    porcentagem = (100 - faltas) / 100 * 100
    return porcentagem

# Função para verificar se o aluno foi aprovado
def verificar_aprovacao(aluno):
    media = calcular_media(aluno)
    porcentagem_presenca = calcular_porcentagem_presenca(aluno)

    if media >= 7 and porcentagem_presenca >= 75:
        return 'Aprovado'
    else:
        return 'Reprovado'

def gerar_csv_aprovados():
  # Cabeçalho do arquivo CSV
  header = ['Nome do Aluno', 'Média', 'Faltas', 'Porcentagem de Presença', 'Situação']

  # Lista para armazenar os dados de cada aluno
  dados_alunos = []

  # Calcular os dados de cada aluno
  for aluno in alunos:
      nome_aluno = aluno['Nome']
      media = calcular_media(aluno)
      faltas = int(aluno['Faltas'])
      porcentagem_faltas = calcular_porcentagem_presenca(aluno)
      situacao = verificar_aprovacao(aluno)
      dados_alunos.append([nome_aluno, round(media, 1), faltas, porcentagem_faltas, situacao])

  # Escrever os dados no arquivo CSV
  with open('aprovacao_alunos.csv', 'w', newline='') as arquivo_csv:
      writer = csv.writer(arquivo_csv)
      writer.writerow(header)
      writer.writerows(dados_alunos)

  print("Arquivo CSV gerado com sucesso!")


# Função para verificar se o aluno fez rematrícula
def verificar_rematricula(aluno):
    return aluno['Rematrícula'] == 'Sim'

# Contadores para a quantidade de alunos aprovados, reprovados, que fizeram rematrícula e os grupos específicos
def exibir_apr_rep ():
  quantidade_aprovados = 0
  quantidade_reprovados = 0
  quantidade_rematricula = 0
  quantidade_aprovados_com_rematricula = 0
  quantidade_aprovados_sem_rematricula = 0
  quantidade_reprovados_com_rematricula = 0
  quantidade_reprovados_sem_rematricula = 0

  # Verificar a situação e rematrícula de cada aluno
  for aluno in alunos:
      situacao = verificar_aprovacao(aluno)
      if situacao == 'Aprovado':
          quantidade_aprovados += 1
          if verificar_rematricula(aluno):
              quantidade_rematricula += 1
              quantidade_aprovados_com_rematricula += 1
          else:
              quantidade_aprovados_sem_rematricula += 1
      else:
          quantidade_reprovados += 1
          if verificar_rematricula(aluno):
              quantidade_rematricula += 1
              quantidade_reprovados_com_rematricula += 1
          else:
              quantidade_reprovados_sem_rematricula += 1

  # Calcular as porcentagens
  porcentagem_aprovados = (quantidade_aprovados / len(alunos)) * 100
  porcentagem_reprovados = (quantidade_reprovados / len(alunos)) * 100
  porcentagem_rematricula = (quantidade_rematricula / len(alunos)) * 100
  porcentagem_aprovados_com_rematricula = (quantidade_aprovados_com_rematricula / quantidade_aprovados) * 100
  porcentagem_aprovados_sem_rematricula = (quantidade_aprovados_sem_rematricula / quantidade_aprovados) * 100
  porcentagem_reprovados_com_rematricula = (quantidade_reprovados_com_rematricula / quantidade_reprovados) * 100
  porcentagem_reprovados_sem_rematricula = (quantidade_reprovados_sem_rematricula / quantidade_reprovados) * 100

  # Exibir os resultados
  print("Quantidade de Alunos Aprovados:", quantidade_aprovados)
  print("Quantidade de Alunos Reprovados:", quantidade_reprovados)
  print("Quantidade de Alunos que Fizeram Rematrícula:", quantidade_rematricula)
  print("Quantidade de Alunos Aprovados que Fizeram Rematrícula:", quantidade_aprovados_com_rematricula)
  print("Quantidade de Alunos Aprovados que Não Fizeram Rematrícula:", quantidade_aprovados_sem_rematricula)
  print("Quantidade de Alunos Reprovados que Fizeram Rematrícula:", quantidade_reprovados_com_rematricula)
  print("Quantidade de Alunos Reprovados que Não Fizeram Rematrícula:", quantidade_reprovados_sem_rematricula)
  print(f"Porcentagem de Alunos que Fizeram Rematrícula: {porcentagem_rematricula}%")
  print(f"Porcentagem de Alunos Aprovados: {porcentagem_aprovados}%")
  print(f"Porcentagem de Alunos Reprovados: {porcentagem_reprovados}%")
  print(f"Porcentagem de Alunos Aprovados que Fizeram Rematrícula: {porcentagem_aprovados_com_rematricula}%")
  print(f"Porcentagem de Alunos Aprovados que Não Fizeram Rematrícula: {porcentagem_aprovados_sem_rematricula}%")
  print(f"Porcentagem de Alunos Reprovados que Fizeram Rematrícula: {porcentagem_reprovados_com_rematricula}%")
  print(f"Porcentagem de Alunos Reprovados que Não Fizeram Rematrícula: {porcentagem_reprovados_sem_rematricula}%")


def atualizar_aluno(aluno, campo, novo_valor):
    # Verificar se o campo existe no dicionário do aluno
    if campo in aluno:
        aluno[campo] = novo_valor
        print(f"Dado '{campo}' do aluno '{aluno['Nome']}' atualizado com sucesso.")
    else:
        print(f"Dado '{campo}' não encontrado para o aluno '{aluno['Nome']}'.")

def inserir_media(aluno):
    aluno['Média'] = calcular_media(aluno)

# Função para buscar um aluno pelo nome ou número de matrícula
def buscar_aluno(opcao, alunos):
    for aluno in alunos:
        if aluno['Nome'] == opcao or aluno['Matrícula'] == opcao:
            return aluno
    return None


def atualizar_dado_csv():
  # Ler o arquivo CSV inicial
  with open('alunos_matric.csv', 'r') as file:
      reader = csv.DictReader(file)
      alunos = list(reader)

  # Realizar a atualização de um aluno específico
  try:
      opcao = input('Informe o nome ou número de matrícula do aluno que deseja encontrar: ')

      aluno_atualizado = buscar_aluno(opcao, alunos)
      if aluno_atualizado:
          campo = input('Informe o campo que deseja atualizar: ')
          novo_valor = input('Informe o novo valor: ')

          atualizar_aluno(aluno_atualizado, campo, novo_valor)

          # Escrever as alterações no arquivo CSV
          with open('alunos.csv', 'w', newline='') as file:
              fieldnames = alunos[0].keys()
              writer = csv.DictWriter(file, fieldnames=fieldnames)
              writer.writeheader()
              writer.writerows(alunos)

          print("Dados do aluno atualizados com sucesso!")
      else:
          print("Aluno não encontrado.")
  except Exception as e:
      print("Ocorreu um erro:", str(e))

def add_colum_media(nome_aluno):
  aluno_atualizado = buscar_aluno(nome_aluno, alunos)
  if aluno_atualizado:
      atualizar_aluno(aluno_atualizado)

  # Escrever as alterações no arquivo CSV
  with open('alunos.csv', 'w', newline='') as file:
      fieldnames = alunos[0].keys()
      writer = csv.DictWriter(file, fieldnames=fieldnames)
      writer.writeheader()
      writer.writerows(alunos)

  print("Dados do aluno atualizados com sucesso!")


def menu(**opcao):
  while True:
    menu_op = ('[1] Exibir todos os alunos\n[2] Procurar um aluno','[3] Exibir notas(de um aluno ou de todos)','[4] Exibir médias','[5] Extrato de pagamento das mensalidades','[6] Aprovação e rematrícula', '[7] Atualizar dados', '[0] Sair')
    for i in menu_op: print(i)

    opcao=  int(input('\nDigite o número referente à opção que deseja:'))

    if opcao == 0:
      break

    elif opcao ==1:
      exibir_dados()

    elif opcao == 2:
      matricula = input('Informe número de matrícula do aluno que deseja encontrar: ')
      pesquisar_aluno(matricula)

    elif opcao == 3:
      print('[1] Exibir todas as notas\n[2] Exibir os alunos com as maiores e menores notas em cada prova\n[3] Exibir notas de um aluno')
      op_nota = int(input('\nDigite o número referente à opção que deseja:'))

      if op_nota ==1:
        exibir_todas_notas()

      elif op_nota ==2:
        alunos_ranking_provas()

      elif op_nota == 3:
        opcao_nota = input('Informe o nome ou número de matrícula do aluno que deseja encontrar: ').capitalize()
        encontrar_nota_aluno(opcao_nota)

    elif opcao == 4:
      print('[1] Exibir todas as médias\n[2] Exibir os alunos com as maiores e menores médis por curso e geral\n[3] Exibir alunos com as mesmas médias\n')
      op_media = int(input('\nDigite o número referente à opção que deseja:'))

      if op_media == 1:
        exibir_todas_medias()

      elif op_media== 2:
        encontrar_medias()

      elif op_media== 3:
        medias_iguais()

    elif opcao == 5:
      arquivo_financeiro()

    elif opcao == 6:
      print('[1] Gerar arquivo CSV com os dados de aprovação\n[2] Exibir por quantidade e porcentagem os aprovados, reprovados e se fizeram ou não rematrícula\n')
      op_nota = int(input('\nDigite o número referente à opção que deseja:'))

      if op_media ==1:
        gerar_csv_aprovados()

      elif op_nota ==2:
        exibir_apr_rep ()

    elif opcao == 7:
      atualizar_dado_csv()
    else:
      continue


menu()
```
